Looking at subject we want to exclude due to non-informative data:
- model can't capture their data (high overestimation in g2w) due to weird performance in ng2w but no difference in outcome probabilites, nor abnormalities in stimuli contigencies
- non-informative data differ between groups
Looking at statistics in included data at 2 weeks and 6 weeks separately (Ttest)

```python
import scipy.io
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from plotting import Plotting
from loading_preparing_data import Panda
```

```python
# load raw data
filepath = '/Users/jolandamalamud/phd/projects/gng_panda_antler/gng_panda/data/'
panda = Panda(filepath)
D = panda.load_data()
# prep action choices to plot
data = panda.extract_data(D)
```

```python
# plotting options
p = Plotting()
sns.set(style="whitegrid")
opt_plot = {"xlabel": 'trial',
            "ylabel": 'probability to go',
            "title": ['go to win', 'go to avoid', 'nogo to win', 'nogo to avoid'],
            "format": False}
```

```python
data_incl = panda.extract_data(D[data['sess']==2])
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['ll2b2a2epxb', 'llb'])
ex = np.nanmean(data_incl['a_correct'],axis=(0,1)) < 0.5
surrogate_data = panda.load_surrogate_data('modelling_results/sess2/', 'll2b2a2epxb')
data_incl = panda.extract_surrogate_data(data_incl, surrogate_data)
data_to_plot = data_incl['a_go']
opt_plot["surr_to_plot"] = data_incl['surr']
opt_plot["excluding"] = (ex | modelling['excluding'])
opt_plot['legend'] = ['empirical', 'model']
p.error_shaded_line_plot(data_to_plot, opt_plot)
print('excluding: ' + str(round(sum(opt_plot["excluding"])/len(opt_plot["excluding"]),2)) \
      + '% of N(at 2 weeks) = ' + str(round(len(opt_plot["excluding"]),2)))
```

```python
# compare exclusion criteria between groups
dfRCT = panda.load_rctdata()
rctids = dfRCT['identifier_n']
group = dfRCT['group']
taskids = np.unique(data_incl['subids'])
group_matched = panda.match_ids(group, taskids, rctids)
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['ll2b2a2epxb', 'llb'])
ex_criteria = [(ex | modelling['excluding']), ex, modelling['excluding']]
ex_criteria_label = ['.', ' due to bad performance (<0.5 mean accuracy).', ' due to random behaviour.']
for i in range(3):
    ex_combined = ex_criteria[i]
    sert = ex_combined[group_matched == 2]
    placebo = ex_combined[group_matched == 1]
    print(scipy.stats.ttest_ind(sert, placebo, equal_var=False))
    print('significantly more people in the sertraline had to be excluded' \
          + ex_criteria_label[i] + '\nSertraline: ' + str(round(np.nansum(sert)/sum(group==2),2)) \
          + '%, placebo: ' + str(round(np.nansum(placebo)/sum(group==1),2)) + '%')
```

```python
missing_task = ~np.in1d(rctids,taskids)
sert = missing_task[group == 2]
placebo = missing_task[group == 1]
print(scipy.stats.ttest_ind(sert, placebo, equal_var=False))
print('no task data;' + ' Sertraline: ' + str(round(np.nansum(sert)/sum(group==2),2)) \
          + '%, placebo: ' + str(round(np.nansum(placebo)/sum(group==1),2)) + '%')
```

```python
# behaviour of excluded participants differ between groups?
opt_plot['group'] = group_matched
opt_plot["excluding"] = ~(ex | modelling['excluding'])
opt_plot["legend"] = ['placebo empirical', 'placebo model','sertraline empirical','sertraline model']
p.group_line_plot(data_to_plot, opt_plot)
print('N(excluded, 2 weeks) = ' + str(sum((ex | modelling['excluding']))) )
```

```python
# parameter (mean and SD) for excluded participants
for i in range(len(modelling['emmap'][:,0])):
    print('mean (SD): ' + str(round(np.nanmean(modelling['emmap'][i,(ex | modelling['excluding'])]),2)) + \
          ', (' + str(round(np.nanstd(modelling['emmap'][i,(ex | modelling['excluding'])]),2)) + ')')
```

Weirdly high outcome sensitivity in reward. Low learning rates and nogo bias make sense. Low noise...

```python
nn, dd = panda.calculate_outcome_probabilities(data_incl, (ex | modelling['excluding']))
print('% of rewards if correct response: ' + str(np.nanmean(nn/dd,axis=0)))
nn, dd = panda.calculate_outcome_probabilities(data_incl, ~(ex | modelling['excluding']))
print('% of rewards if correct response: ' + str(np.nanmean(nn/dd,axis=0)))
nn, dd = panda.calculate_outcome_probabilities(data_incl, np.full([np.shape(data_incl['a'])[1]], True))
print('% of rewards if correct response: ' + str(np.nanmean(nn/dd,axis=0)))
```

```python
print('How much do the exluding criteria overlap: ' + str(round(sum(ex == modelling['excluding'])/len(ex),3)))
```

```python
# particpants who are NOT better captured by the random model but perform very poorly 
opt_plot["excluding"] = ~(~modelling['excluding'] & ex)
opt_plot['legend'] = ['empirical', 'model']
p.error_shaded_line_plot(data_to_plot, opt_plot)
print('N = ' + str(sum((~modelling['excluding'] & ex))) )
```

```python
nn, dd = panda.calculate_outcome_probabilities(data_incl, (~modelling['excluding'] & ex))
print('% of rewards: ' + str(np.nanmean(nn/24,axis=0)))
print('% of correct response: ' + str(np.nanmean(dd,axis=0)/24))
print('% of rewards if correct response: ' + str(np.nanmean(nn/dd,axis=0)))
```

```python
# stimuli presentations in first few trials 
T = 5
ex_criteria = [(~modelling['excluding'] & ex), (modelling['excluding'] | ex), ~(modelling['excluding'] | ex)]
ex_criteria_label = [' pressing more in ng2w than g2w', ' excluded.', ' included.']
for i in range(3):
    ex_combined = ex_criteria[i]
    print('people which are' + ex_criteria_label[i])
    for i in range(4):
        print(round(sum(sum(data_incl['stimuli'][0:T,ex_combined] == i+1)) \
              / (sum(ex_combined)*T),2))
```

```python
# particpants who are better captured by the random model but do NOT perform less than 0.5 accuracy
opt_plot["excluding"] = ~(modelling['excluding'] & ~ex)
opt_plot['legend'] = ['empirical', 'model']
p.error_shaded_line_plot(data_to_plot, opt_plot)
print('N = ' + str(sum((modelling['excluding'] & ~ex))) )
```

```python
# compare parameter in group
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['ll2b2a2epxb', 'llb'])
sert = modelling["emmap"][:,(~(ex | modelling['excluding']) & (group_matched == 2))]
placebo = modelling["emmap"][:,(~(ex | modelling['excluding']) & (group_matched == 1))]
print('mean and SD of cognitive parameters and pvalue (2sample ttest)')
parameter_labels = ['reward sensitivity', 'loss sensitivity', 'reward LR', 'loss LR', 'appetitive Pav', 'aversive Pav', 'noise', 'bias']
panda.ttest2(sert, placebo, parameter_labels)
```
