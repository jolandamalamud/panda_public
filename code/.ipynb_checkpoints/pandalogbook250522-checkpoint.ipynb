```python
import scipy.io
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from plotting import Plotting
from loading_preparing_data import Panda
```

```python
# load raw data
filepath = '/Users/jolandamalamud/phd/projects/gng_panda_antler/gng_panda/data/'
panda = Panda(filepath)
D = panda.load_data()
# prep action choices to plot
data = panda.extract_data(D)
```

```python
# plotting options
p = Plotting()
sns.set(style="whitegrid")
opt_plot = {"xlabel": 'trial',
            "ylabel": 'probability to go',
            "title": ['go to win', 'go to avoid', 'nogo to win', 'nogo to avoid'],
            "format": False}
```

```python
data_incl = panda.extract_data(D[data['sess']==2])
accuracy_better50 = np.nanmean(data_incl['a_correct'],axis=(0,1)) > 0.5
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['ll2b2a2epxb', 'llb'])
exclusion = modelling['excluding'] | (~accuracy_better50)
```

```python
# plotting surrogate data from the better fitting model 
surrogate_data_in = panda.load_surrogate_data('modelling_results/sess2/', 'll2b2a2epxb')
surrogate_data_ex = panda.load_surrogate_data('modelling_results/sess2/', 'llb')
opt_plot["surr_to_plot"] =  np.full_like(data_incl['a_go'], np.nan)
data_incl = panda.extract_surrogate_data(data_incl, surrogate_data_in)
opt_plot["surr_to_plot"][:,:,~modelling['excluding']] = data_incl['surr'][:,:,~modelling['excluding']]
data_incl = panda.extract_surrogate_data(data_incl, surrogate_data_ex)
opt_plot["surr_to_plot"][:,:,modelling['excluding']] = data_incl['surr'][:,:,modelling['excluding']]
data_incl['surr'] = opt_plot["surr_to_plot"]
```

```python
data_to_plot = data_incl['a_go']
opt_plot['legend'] = ['empirical', 'model']
p.error_shaded_line_plot(data_to_plot, opt_plot)
p.goprobability_scatter_plot(data_incl, opt_plot)
print('N(most parsimonious model) = ' + str(sum(~modelling['excluding'])) + ', N(random baseline model) = ' + str(sum(modelling['excluding'])))
```

```python
surrogate_data = panda.load_surrogate_data('modelling_results/sess2/', 'll2b2a2epxb')
opt_plot["surr_to_plot"] =  np.full_like(data_incl['a_go'], np.nan)
data_incl = panda.extract_surrogate_data(data_incl, surrogate_data)
opt_plot["surr_to_plot"] = data_incl['surr']
opt_plot['excluding'] = exclusion
opt_plot['legend'] = ['empirical', 'model']
p.error_shaded_line_plot(data_to_plot, opt_plot)
p.goprobability_scatter_plot(data_incl, opt_plot)
print('N(included) = ' + str(sum(~opt_plot['excluding'])))
```

```python
# compare exclusion criteria between groups
dfRCT = panda.load_rctdata()
rctids = dfRCT['identifier_n']
group = dfRCT['group']
taskids = np.unique(data_incl['subids'])
group_matched = panda.match_ids(group, taskids, rctids)
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['ll2b2a2epxb', 'llb'])
ex_criteria = [((~accuracy_better50) | modelling['excluding']), (~accuracy_better50), modelling['excluding']]
ex_criteria_label = ['.', ' due to bad performance (<=0.5 mean accuracy).', ' due to random behaviour.']
for i in range(3):
    ex_combined = ex_criteria[i]
    sert = ex_combined[group_matched == 2]
    placebo = ex_combined[group_matched == 1]
    print(scipy.stats.ttest_ind(sert, placebo, equal_var=False))
    print('significantly more people in the sertraline group had to be excluded' \
          + ex_criteria_label[i] + '\nSertraline: ' + str(round(np.nansum(sert)/sum(group==2),2)) \
          + '%, placebo: ' + str(round(np.nansum(placebo)/sum(group==1),2)) + '%')
```

```python
missing_task = ~np.in1d(rctids,taskids)
sert = missing_task[group == 2]
placebo = missing_task[group == 1]
print(scipy.stats.ttest_ind(sert, placebo, equal_var=False))
print('no task data;' + ' Sertraline: ' + str(round(np.nansum(sert)/sum(group==2),2)) \
          + '%, placebo: ' + str(round(np.nansum(placebo)/sum(group==1),2)) + '%')
```

```python
data_incl = panda.extract_data(D[(data['sess']==1) | (data['sess']==2)])
taskids = data_incl['subids']
dfRCT = panda.load_rctdata()
rctids = dfRCT['identifier_n']
rct_variables = dfRCT[{'group','_site_n','cistotal_cat','depr_dur_2years'}]
rct_variables_matched = pd.DataFrame()
for c in rct_variables.columns:
    rct_variables_matched[c] = panda.match_ids(rct_variables[c], taskids, rctids)
```

```python
modelling = panda.load_modelling_results('/modelling_results/sess12/', ['ll2b2a2epxb', 'llb'])
accuracy_better50 = np.nanmean(data_incl['a_correct'],axis=(0,1)) >= 0.5
exclusion = modelling['excluding'] | (~accuracy_better50)
```

```python
subjects = np.full([len(data_incl['subids'])], np.nan)
for i, subid in enumerate(np.unique(data_incl['subids'])):
    subjects[np.where(data_incl['subids'] == subid)] = (i + 1)
    
group_all = np.hstack((np.zeros(np.sum(data_incl['sess']==1)), rct_variables_matched['group'][data_incl['sess']==2]))
time = np.hstack((np.zeros(np.sum(data_incl['sess']==1)), 2*np.ones(np.sum(data_incl['sess']==2))))
```

```python
# using statsmodel package for LME (drop all data if nan in any row)
import statsmodels.formula.api as smf
parameter_labels = ['reward sensitivity', 'loss sensitivity', 'reward LR', 'loss LR', 'appetitive Pav', 'aversive Pav', 'noise', 'bias']
for i in range(np.shape(modelling["emmap"])[0]):
    parameter = modelling["emmap"][i,:]
    df = pd.DataFrame({'subject': subjects, 'parameter': parameter,\
                   'group': group_all, 'time': time, 'site': rct_variables_matched['_site_n'], \
                      'cis': rct_variables_matched['cistotal_cat'], 'dep': rct_variables_matched['depr_dur_2years']})
    df = df[~exclusion]
    df = df.dropna()
    model = smf.mixedlm("parameter ~ group + time + site + dep + cis", df, groups= 'subject').fit()
    print('####################')
    print(parameter_labels[i])
    print('####################')
    print(model.summary())
```

```python
random_baseline = panda.load_modelling_results('/modelling_results/sess12/', ['llb'])
exclusion = ~modelling['excluding'] | (~accuracy_better50)
parameter = random_baseline["emmap"][0]
df = pd.DataFrame({'subject': subjects, 'parameter': parameter,\
               'group': group_all, 'time': time, 'site': rct_variables_matched['_site_n'], \
                  'cis': rct_variables_matched['cistotal_cat'], 'dep': rct_variables_matched['depr_dur_2years']})
df = df[~exclusion]
df = df.dropna()
model = smf.mixedlm("parameter ~ group + time + site + dep + cis", df, groups= 'subject').fit()
model.summary()
```

```python
# using R Lmer package for LME
from pymer4.models import Lmer
parameter_labels = ['reward sensitivity', 'loss sensitivity', 'reward LR', 'loss LR', 'appetitive Pav', 'aversive Pav', 'noise', 'bias']
for i in range(np.shape(modelling["emmap"])[0]):
    parameter = modelling["emmap"][i,:]
    df = pd.DataFrame({'subject': subjects, 'parameter': parameter,\
                       'group': group_all, 'time': time})
    df = df[~exclusion]
    model = Lmer("parameter  ~ group + time + (1|subject)", data=df).fit();
    print('####################')
    print(parameter_labels[i])
    print('####################')
    print(model)
```
