```python
import scipy.io
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from plotting import Plotting
from loading_preparing_data import Panda
import statsmodels.formula.api as smf
```

```python
# load raw data
filepath = '/Users/jolandamalamud/phd/projects/gng_panda_antler/gng_panda/data/'
panda = Panda(filepath)
D = panda.load_data()
# prep action choices to plot
data = panda.extract_data(D)
```

```python
# plotting options
p = Plotting()
sns.set(style="whitegrid")
opt_plot = {"xlabel": 'trial',
            "ylabel": 'probability to go',
            "title": ['go to win', 'go to avoid', 'nogo to win', 'nogo to avoid'],
            "format": False}
```

```python
# first we only look at session 2 (2 weeks) as there a drug effect is most expected
data_incl = panda.extract_data(D[data['sess']==2])
accuracy_less_50 = np.nanmean(data_incl['a_correct'],axis=(0,1)) < 0.5
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['ll2b2a2epxb', 'llb'])
exclusion = modelling['excluding'] | accuracy_less_50
```

```python
# plotting surrogate data from the better fitting model 
surrogate_data_in = panda.load_surrogate_data('modelling_results/sess2/', 'll2b2a2epxb')
surrogate_data_ex = panda.load_surrogate_data('modelling_results/sess2/', 'llb')
opt_plot["surr_to_plot"] =  np.full_like(data_incl['a_go'], np.nan)
data_incl = panda.extract_surrogate_data(data_incl, surrogate_data_in)
opt_plot["surr_to_plot"][:,:,~modelling['excluding']] = data_incl['surr'][:,:,~modelling['excluding']]
data_incl = panda.extract_surrogate_data(data_incl, surrogate_data_ex)
opt_plot["surr_to_plot"][:,:,modelling['excluding']] = data_incl['surr'][:,:,modelling['excluding']]
data_incl['surr'] = opt_plot["surr_to_plot"]
```

```python
data_to_plot = data_incl['a_go']
opt_plot['legend'] = ['empirical', 'model']
p.error_shaded_line_plot(data_to_plot, opt_plot)
p.goprobability_scatter_plot(data_incl, opt_plot)
print('N(most parsimonious model) = ' + str(sum(~modelling['excluding'])) + ', N(random baseline model) = ' + str(sum(modelling['excluding'])))
```

```python
# plotting surrogate data only for most parsiminious model excluding participants which were 
# better fit by random baseline model and/or performed less than 50
surrogate_data = panda.load_surrogate_data('modelling_results/sess2/', 'll2b2a2epxb')
data_incl = panda.extract_surrogate_data(data_incl, surrogate_data)
opt_plot["surr_to_plot"] = data_incl['surr']
opt_plot['excluding'] = exclusion
opt_plot['legend'] = ['empirical', 'model']
p.error_shaded_line_plot(data_to_plot, opt_plot)
p.goprobability_scatter_plot(data_incl, opt_plot)
print('N(included) = ' + str(sum(~exclusion)))
```

```python
# compare exclusion criteria between groups
dfRCT = panda.load_rctdata()
rctids = dfRCT['identifier_n']
group = dfRCT['group']
taskids = np.unique(data_incl['subids'])
group_matched = panda.match_ids(group, taskids, rctids)
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['ll2b2a2epxb', 'llb'])
ex_criteria = [(accuracy_less_50 | modelling['excluding']), accuracy_less_50, modelling['excluding']]
ex_criteria_label = ['.', ' due to bad performance (<=0.5 mean accuracy).', ' due to random behaviour.']
for i in range(3):
    ex_combined = ex_criteria[i]
    sert = ex_combined[group_matched == 2]
    placebo = ex_combined[group_matched == 1]
    print(scipy.stats.ttest_ind(sert, placebo, equal_var=False))
    print('significantly more people in the sertraline group had to be excluded' \
          + ex_criteria_label[i] + '\nSertraline: ' + str(round(np.nansum(sert)/sum(group==2),2)) \
          + '%, placebo: ' + str(round(np.nansum(placebo)/sum(group==1),2)) + '%')
```

```python
missing_task = ~np.in1d(rctids,taskids)
sert = missing_task[group == 2]
placebo = missing_task[group == 1]
print(scipy.stats.ttest_ind(sert, placebo, equal_var=False))
print('no task data;' + ' Sertraline: ' + str(round(np.nansum(sert)/sum(group==2),2)) \
          + '%, placebo: ' + str(round(np.nansum(placebo)/sum(group==1),2)) + '%')
```

```python
# effect of sertraline on anxiety different in included and excluded group?
gad_df = dfRCT[['gadtot','log_gadtot_2wk', 'log_gadtot_6wk', 'log_gadtot_12wk']]
gad_df['log_gadtot'] = np.log10(gad_df['gadtot']+1)
gad_matched = pd.DataFrame()
for c in gad_df:
    gad_matched[c] = panda.match_ids(gad_df[c], taskids, rctids)
exclusion = modelling['excluding'] #(accuracy_less_50 | modelling['excluding'])
performance_split = [exclusion, ~exclusion]
performance_split_label = ['excluded', 'included']
for i in range(len(performance_split)):
    df = pd.DataFrame({'subject': np.arange(len(group_matched)),'group': group_matched, \
                       'gad0': gad_matched['log_gadtot'], 'gad2': gad_matched['log_gadtot_2wk'], \
                       'gad6': gad_matched['log_gadtot_6wk'],'gad12': gad_matched['log_gadtot_12wk']})
    df = df[performance_split[i]]
    df = df.dropna()
    model = smf.mixedlm("gad2 ~ gad0 + group", df, groups= 'subject').fit()
    print(performance_split_label[i] + ' participants at week 2')
    print('-----------------------------------------------------')
    print(model.summary())
    model = smf.mixedlm("gad6 ~ gad0 + group", df, groups= 'subject').fit()
    print(performance_split_label[i] + ' participants at week 6')
    print('-----------------------------------------------------')
    print(model.summary())
    model = smf.mixedlm("gad12 ~ gad0 + group", df, groups= 'subject').fit()
    print(performance_split_label[i] + ' participants at week 6')
    print('-----------------------------------------------------')
    print(model.summary())
```

At two weeks the PANDA study couldn't show an effect of sertraline on anxiety. However, if we split the participant depending on their task behaviour, we see significant improvement in people who couldn't do the task properly at 2 and 6 weeks and no effect of sertraline on anxiety in people who did the task properly, neither at 2 nor 6 weeks! 

```python
# Is the loss LR related to anxiety
parameter = modelling["emmap"][3,:]
df = pd.DataFrame({'subject': np.arange(len(parameter)), 'parameter': parameter,\
                   'group': group_matched, 'gad0': gad_matched['log_gadtot'], \
                   'gad2': gad_matched['log_gadtot_2wk'], \
                   'gad6': gad_matched['log_gadtot_6wk'],'gad12': gad_matched['log_gadtot_12wk']})
df = df[~(accuracy_less_50 | modelling['excluding'])]
df = df.dropna()
model = smf.mixedlm("gad2 ~ gad0 + group + parameter", df, groups= 'subject').fit()
print(model.summary())
model = smf.mixedlm("gad6 ~ gad0 + group + parameter", df, groups= 'subject').fit()
print(model.summary())
model = smf.mixedlm("gad12 ~ gad0 + group + parameter", df, groups= 'subject').fit()
print(model.summary())
```

```python
# using statsmodel package for LME (drop all data if nan in any row)
parameter_labels = ['reward sensitivity', 'loss sensitivity', 'reward LR', 'loss LR', 'appetitive Pav', 'aversive Pav', 'noise', 'bias']
for i in range(np.shape(modelling["emmap"])[0]):
    parameter = modelling["emmap"][i,:]
    df = pd.DataFrame({'subject': subjects, 'parameter': parameter,\
                   'group': group_all, 'time': time, 'site': rct_variables_matched['_site_n'], \
                      'cis': rct_variables_matched['cistotal_cat'], 'dep': rct_variables_matched['depr_dur_2years']})
    df = df[~exclusion]
    df = df.dropna()
    model = smf.mixedlm("parameter ~ group + time + site + dep + cis", df, groups= 'subject').fit()
    print('####################')
    print(parameter_labels[i])
    print('####################')
    print(model.summary())
```
