```python
import scipy.io
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from plotting import Plotting
from loading_preparing_data import Panda
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
# load raw data
filepath = '/Users/jolandamalamud/phd/projects/gng_panda_antler/gng_panda/data/'
panda = Panda(filepath)
D = panda.load_data()
# prep action choices to plot
data = panda.extract_data(D)
Nsj = len(data['subids'])
```

```python
# plotting options
p = Plotting()
sns.set(style="white")
plt.rcParams.update({'font.size': 18})
opt_plot = {"xlabel": 'trial',
            "ylabel": 'probability to go',
            "title": ['go to win', 'go to avoid', 'nogo to win', 'nogo to avoid'],
            "format": False}
figpath = '/Users/jolandamalamud/phd/projects/gng_panda_antler/gng_panda/results/poster_figures/'
```

```python
# plotting surrogate data from the better fitting model 
modelling = panda.load_modelling_results('/modelling_results/final_results/', ['ll2b2a2epxb', 'llb'])
surrogate_data_in = panda.load_surrogate_data('modelling_results/final_results/', 'll2b2a2epxb')
surrogate_data_ex = panda.load_surrogate_data('modelling_results/final_results/', 'llb')
opt_plot["surr_to_plot"] =  np.full_like(data['a_go'], np.nan)
data = panda.extract_surrogate_data(data, surrogate_data_in)
opt_plot["surr_to_plot"][:,:,~modelling['excluding']] = data['surr'][:,:,~modelling['excluding']]
data = panda.extract_surrogate_data(data, surrogate_data_ex)
opt_plot["surr_to_plot"][:,:,modelling['excluding']] = data['surr'][:,:,modelling['excluding']]
data['surr'] = opt_plot["surr_to_plot"]
```

```python
# empirical and surrogate data from better fitting model all sessions collabsed together
import matplotlib.patches as patches
cmapmine = ListedColormap(['lightgrey', 'w'], N=2)
fig, ax = plt.subplots(1,4,figsize=(15, 6))
clrs = ['forestgreen', 'tomato','limegreen','red']
for i in range(4):
    im = ax[i].imshow((data['a_go'][:,i,:]==1).astype(int).T, cmap=cmapmine, aspect='auto',origin='lower')
    epochs = np.arange(np.shape(data['a_go'])[0])
    meanst = np.nanmean(data['a_go'][:,i,:],axis=1)
    sdt = np.nanstd(data['a_go'][:,i,:],axis=1)/np.sqrt(Nsj)
    p, = ax[i].plot(epochs, meanst*Nsj,linewidth=1)
#     ax[i].fill_between(epochs, (meanst - sdt)*Nsj,(meanst + sdt)*Nsj, alpha=0.3)
    ax[i].plot(epochs, np.nanmean(data['surr'][:,i,:],axis=1)*Nsj,color=p.get_color(),linewidth=2,linestyle='--')
    ax[i].tick_params(labelsize=26)
    if i == 0:
        ax[i].set_yticks([1, np.round(Nsj/4), np.round(Nsj/2), np.round(Nsj/4*3), Nsj])
        ax[i].set_yticklabels([0,0.25,0.5,0.75,1])
        ax[i].set_ylabel('go probability', fontsize=32)
    else:
        ax[i].set_yticks([])
    ax[i].set_title(opt_plot['title'][i], fontsize=32 ,weight='bold')
    ax[i].set_xticks([])
    #ax[i].set_xlabel('trials', fontsize=20)
fig.tight_layout()
plt.savefig(figpath + 'surrogatedata.svg', format='svg',transparent=True)
legend1 = plt.legend(['empirical data', 'model data'],fontsize=16,loc=(1.04,0))
values = {'go':1, 'nogo': 0}
colors = [im.cmap(im.norm(value)) for value in values.values()] 
patches = [patches.Patch(facecolor=colors[i],edgecolor='lightgrey', label=list(values.keys())[i]) for i in range(len(values))]
plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.2, fontsize=16)
plt.gca().add_artist(legend1)
plt.savefig(figpath + 'surrogatedata_legend.svg', format='svg',transparent=True,bbox_inches='tight')
```
