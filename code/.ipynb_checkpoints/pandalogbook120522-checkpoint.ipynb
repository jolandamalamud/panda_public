Check more closely where the problem from the most parsimonious model arises. If we exclude people who perform less than 0.5 on average (over all conditions) and who behave randomly (better fitted by random baseline model), model capture behaviour well, but we have to exclude half of the participants:

```python
data_incl = panda.extract_data(D[data["sess"]==2])
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['ll2b2a2epxb', 'llb'])
ex = np.nanmean(data_incl['a_correct'],axis=(0,1)) < 0.5
surrogate_data = panda.load_surrogate_data('modelling_results/sess2/', 'll2b2a2epxb')
data_incl = panda.extract_surrogate_data(data_incl, surrogate_data)
data_to_plot = data_incl['a_go']
opt_plot["surr_to_plot"] = data_incl['surr']
opt_plot["excluding"] = (ex | modelling['excluding'])
opt_plot['legend'] = ['empirical', 'model']
p.error_shaded_line_plot(data_to_plot, opt_plot)
p.goprobability_scatter_plot(data_incl, opt_plot)
print('excluding: ' + str(round(sum(opt_plot["excluding"])/len(opt_plot["excluding"]),2)) \
      + '% of N(at 2 weeks) = ' + str(round(len(opt_plot["excluding"]),2)))
```

```python
print('How much do the exluding criteria overlap: ' + str(round(sum(ex == modelling['excluding'])/len(ex),3)))
```

```python
opt_plot["excluding"] = ~(~modelling['excluding'] & ex)
opt_plot['legend'] = ['empirical', 'model']
p.error_shaded_line_plot(data_to_plot, opt_plot)
```

```python
# clustering performance to see where the model misfit takes place
from sklearn.cluster import KMeans
kmeans = KMeans(init="random",n_clusters=2,n_init=10,max_iter=300,random_state=42)
k = kmeans.fit(np.nanmean(data_incl['a_correct'],axis=0).T)
```

```python
for i in range(2):
    opt_plot["excluding"] = ~(k.labels_ == i)
    opt_plot['legend'] = ['empirical', 'model']
    p.error_shaded_line_plot(data_to_plot, opt_plot)
    plt.suptitle('cluster ' + str(i+1) + ' N(' + str(sum(k.labels_ == i)) + ')', size=18);
```

Problem is mainly in the people who very rarely press... Why is the model not able to capture this?

Performing statistics on the participants which did the task well (~50% excluded due to less than 0.5 overall accuracy and better explained by the complex model than the random baseline model) using the parameter from ll2b2a2epxb comparing groups in follow-up 1 (2 weeks)

```python
# trial data
dfRCT = panda.load_rctdata()
rctids = dfRCT['identifier_n']
group = dfRCT['group']
taskids = np.unique(data_incl['subids'])
group_matched = panda.match_ids(group, taskids, rctids)
```

```python
# compare parameter in group
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['ll2b2a2epxb', 'llb'])
sert = modelling["emmap"][:,(~(ex | modelling['excluding']) & (group_matched == 2))]
placebo = modelling["emmap"][:,(~(ex | modelling['excluding']) & (group_matched == 1))]
print('mean and SD of cognitive parameters and pvalue (2sample ttest)')
parameter_labels = ['reward sensitivity', 'loss sensitivity', 'reward LR', 'loss LR', 'appetitive Pav', 'aversive Pav', 'noise', 'bias']
panda.ttest2(sert, placebo, parameter_labels)
```

```python
# relation to psychiatric score
weeks = ['2', '6', '12']
for i in weeks:
    gad = dfRCT['log_gadtot_'+i+'wk']
    gad_matched = panda.match_ids(gad, taskids, rctids)
    print('association between...')
    panda.correlation(modelling['emmap'][:,~modelling['excluding']], \
                      gad_matched[~modelling['excluding']], parameter_labels, 'total gad at '+i+ ' weeks')
```

No relationship with current or future GAD-7 score. But will further look into if changes in parameters are related to anxiety. Sertraline group shows significant higher reward sensitivity and loss learning rate, and a trend towards lower aversive Pavlovian bias. Similar results when looking at the parameters of winning model where the ng2w condition has a separate Pavlovian bias.

```python
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['ll2b2a2epxb20', 'llb'])
sert = modelling["emmap"][:,(~(ex | modelling['excluding']) & (group_matched == 2))]
placebo = modelling["emmap"][:,(~(ex | modelling['excluding']) & (group_matched == 1))]
print('mean and SD of cognitive parameters and pvalue (2sample ttest)')
parameter_labels = ['reward sensitivity', 'loss sensitivity', 'reward LR', 'loss LR', 'ng2w Pav', 'Pav', 'noise', 'bias']
panda.ttest2(sert, placebo, parameter_labels)
```
