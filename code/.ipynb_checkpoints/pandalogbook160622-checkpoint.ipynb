```python
exclusion = []
for t in range(number_of_sessions):
    exclusion.append(np.nansum(df_panda['exclusion' + str(t+1)]))
    print('# of excluded subjects at week ' + str(weeks[t]) + ':\t' + str(exclusion[t]) + \
      ' (' + str(round(exclusion[t]/number_of_subjects*100)) +'%)')
print('total # of excluded subjects:\t\t' + str(sum(exclusion)) + \
      ' (' + str(round(sum(exclusion)/(number_of_sessions*number_of_subjects)*100)) +'%)')
```

```python
# chi2 pearson test to compare included vs excluded in sertraline vs placebo
tmp = panda.chi2_test(df_panda['exclusion2'], df_panda['group'], ['sertraline', 'placebo'])
tmp.insert(0, 'variable', 'exclusion')
display(tmp)
print('N = ' + str(sum(~df_panda['exclusion2'].isna() & ~df_panda['group'].isna())) + \
      ', sertraline = ' + str(np.round(df_panda['exclusion2'][df_panda['group']==1].mean()*100, 2)) + '%'\
      + ', placebo = ' + str(np.round(df_panda['exclusion2'][df_panda['group']==0].mean()*100, 2)) + '%')
```

```python
# is exclusion related to baseline variables?
baseline_continuous =  ['gad1log', 'phq1log', 'age', 'cisscore']
baseline_categorical = ['site', 'cis', 'dep', 'education', \
                        'AD_past','sex', 'ethnic', 'fin', 'empstat', 'marstat']
tab_list = []
for t in range(1,number_of_sessions+1):
    tab = pd.DataFrame()
    for i in baseline_continuous:
        tmp = panda.group_ttest(df_panda[i], df_panda['exclusion' + str(t)], ['excluded', 'included'])
        tmp.insert(0, 'variable', i)
        tab = pd.concat((tab, tmp))
    for i in baseline_categorical:
        tmp = panda.chi2_test(df_panda[i], df_panda['exclusion'  + str(t)], ['excluded', 'included'])
        tmp.insert(0, 'variable', i)
        tab = pd.concat((tab, tmp))
    tab_list.append(tab)
disp.display_side_by_side(tab_list[0],tab_list[1])
```

---------------------------------------------------------------------------------------------------------------
MIXED EFFECTS MODELLING
--------------------------------------------------------------------------------------------------------

```python
# create dataframe for mixed-effects modelling
mle_df = pd.DataFrame(columns = gng_columns + rct_columns)
for c in gng_columns:
    if number_of_sessions == 2:
        mle_df[c] = pd.concat([df_panda[c + '1'], df_panda[c + '2']], axis=0, ignore_index=True)
    elif number_of_sessions == 3:
        mle_df[c] = pd.concat([df_panda[c + '1'], df_panda[c + '2'], df_panda[c + '3']], axis=0, ignore_index=True)
    
for c in rct_columns:
    mle_df[c] = pd.concat([df_panda[c], df_panda[c]], axis=0, ignore_index=True)


mle_df['subject'] = np.hstack((np.arange(len(df_panda)),np.arange(len(df_panda))))
mle_df['time'] = np.hstack((np.tile(1, len(df_panda)),np.tile(2, len(df_panda))))
mle_df['weeks'] = np.hstack((np.tile(weeks[0], len(df_panda)),np.tile(weeks[1], len(df_panda))))
mle_df['group'] = np.hstack((np.tile(0, len(df_panda)),df_panda['group']))
mle_df['gadlog'] = np.hstack((df_panda['gad1log'],df_panda['gad2log']))
```

```python
# does exlusion criteria have an effect on the effect of sertraline on anxiety?
mle_df_group_exclusion_interaction = mle_df.copy()
for i in range(2,5):
    mle_df_group_exclusion_interaction['gadlog'] = np.hstack((df_panda['gad1log'], df_panda['gad' + str(i) + 'log']))
    model = smf.mixedlm('gadlog ~ group * exclusion + time + site + cis + dep', \
                        mle_df_group_exclusion_interaction, groups= 'subject', missing='drop').fit()
    print(model.summary())
```

```python
# Post-hoc influence of group on anxiety in excluded and included
mle_df_group_exclusion_interaction['gadlog'] = np.hstack((df_panda['gad1log'], df_panda['gad3log']))
for i in range(2):
    model = smf.mixedlm('gadlog ~ group + time + site + cis + dep + age + education + AD_past + marstat', \
                        mle_df_group_exclusion_interaction[mle_df_group_exclusion_interaction['exclusion']==i], groups= 'subject', missing='drop').fit()
    print(model.summary())
```

```python
# does the group have an effect on accuracy when controlling for the exclusion group?
x = [x for x in gng_columns if 'acc' in x]
for i in x:
    model = smf.mixedlm(i+' ~ group*exclusion + site + cis + dep + time', \
                    mle_df, groups= 'subject', missing='drop').fit()
    print(model.summary())
```

```python
# does task performance have an effect on gad at 6 weeks controlled for the exclusion and drug group?
mle_df['gadlog'] = np.hstack((df_panda['gad1log'], df_panda['gad3log']))
x = [x for x in gng_columns if 'acc' in x]
for i in x:
    model = smf.mixedlm('gadlog ~ '+i+' * exclusion * group + time + age + education + sex', \
                    mle_df, groups= 'subject', re_formula='~'+i+'', missing='drop').fit()
    print(model.summary())
```

```python
# do cognitive parameters differ between drug group controlling for exclusion group?
for i in parameter_labels:
    model = smf.mixedlm(i + ' ~ group * exclusion + time + site + cis + dep', \
                        mle_df, groups= 'subject', missing='drop').fit()
    print(model.summary())
```

```python
# do cognitive parameters have an effect on gad at 6 weeks controlled for the exclusion and drug group?
mle_df['gadlog'] = np.hstack((df_panda['gad1log'], df_panda['gad3log']))
for i in parameter_labels:
    model = smf.mixedlm('gadlog ~ ' + i + '*exclusion * group + time + age + education + sex', \
                        mle_df, groups= 'subject', re_formula='~' + i, missing='drop').fit()
    print(model.summary())
```

```python
# do cognitive parameters differ between drug group in the included group?
for i in parameter_labels[:-1]:
    model = smf.mixedlm(i + ' ~ group + time + site + cis + dep + age + education + AD_past + marstat', \
                        mle_df[mle_df['exclusion']==0], groups= 'subject', missing='drop').fit()
    print(model.summary())
```

```python
# does the random bias differ between drug group in the excluded group?
model = smf.mixedlm('rbias ~ group + time + site + cis + dep + age + education + AD_past + marstat', \
                    mle_df[mle_df['exclusion']==1], groups= 'subject', missing='drop').fit()
print(model.summary())
```

```python
# do cognitive parameters have an effect on gad at 6 weeks in the included group controlled drug group?
mle_df['gadlog'] = np.hstack((df_panda['gad1log'], df_panda['gad3log']))
for i in parameter_labels[:-1]:
    model = smf.mixedlm('gadlog ~ ' + i + ' + group + time + age + education + sex + AD_past', \
                        mle_df[mle_df['exclusion']==0], groups= 'subject', re_formula='~' + i, missing='drop').fit()
    print(model.summary())
```

```python
# do rbias have an effect on gad at 6 weeks in the excluded group controlled drug group?
mle_df['gadlog'] = np.hstack((df_panda['gad1log'], df_panda['gad3log']))
model = smf.mixedlm('gadlog ~ rbias + group + time + age + education + sex + AD_past', \
                    mle_df[mle_df['exclusion']==1], groups= 'subject', re_formula='~rbias', missing='drop').fit()
print(model.summary())
```
