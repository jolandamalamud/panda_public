```python
import scipy.io
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.io
```

```python
filepath = '/Users/jolandamalamud/phd/projects/gng_panda_antler/gng_panda/'
mat = scipy.io.loadmat(filepath + 'results/modelling/recovery/ll2b2a2epxb_recovery_202208220000.mat')
restart = 100
n_parameters = 8
Nsj = 200
parameter_labels = ['reward\nsensitivity', 'loss\nsensitivity', 'reward\nlearning rate', \
                    'loss\nlearning rate', 'appetitive\nPavlovian bias', 'aversive\nPavlovian bias', \
                    'noise', 'bias']
```

```python
if restart == 1:
    parameter_estimates = np.empty([Nsj, n_parameters])
    parameter_true = np.empty([Nsj, n_parameters])
    for j in range(n_parameters):
        parameter_estimates[:,j] = np.array(mat['fit']['parest'][0][0][j])
    parameter_true = np.hstack(mat['Data']['trueParam'][0]).T
    
    corrmatrix = np.empty([n_parameters,n_parameters])
    for i in range(n_parameters):
        for j in range(n_parameters):
            corrmatrix[i,j] = np.corrcoef(parameter_estimates[:,i], \
                                          parameter_true[:,j])[1,0]
```

```python
if restart > 1:
    parameter_estimates = np.empty([restart, Nsj, n_parameters])
    parameter_true = np.empty([restart, Nsj, n_parameters])
    for j in range(n_parameters):
        tmp = []
        for i in range(restart):
            parameter_true[i,:,:] = np.hstack(mat['D'][0][i]['trueParam'][0]).T
            tmp.append(mat['fit'][0][i]['parest'][0][0][j])
        parameter_estimates[:,:,j] = np.array(tmp)
    
    corrmatrix = np.empty([n_parameters,n_parameters])
    for i in range(n_parameters):
        for j in range(n_parameters):
            corrmatrix[i,j] = np.corrcoef(np.nanmean(parameter_estimates,axis=(0))[:,i], \
                                          np.nanmean(parameter_true,axis=(0))[:,j])[1,0]
```

```python
if restart > 1:
    parameter_estimates = np.empty([restart, Nsj, n_parameters])
    parameter_true = np.empty([restart, Nsj, n_parameters])
    corrmatrix = np.empty([n_parameters, restart])
    for j in range(n_parameters):
        tmp = []
        for i in range(restart):
            parameter_true[i,:,:] = np.hstack(mat['D'][0][i]['trueParam'][0]).T
            parameter_estimates[i,:,j] = mat['fit'][0][i]['parest'][0][0][j]
```

```python
corrmatrix = np.empty([n_parameters,n_parameters,restart])
for f in range(restart):
    for i in range(n_parameters):
        for j in range(n_parameters):
            corrmatrix[i,j,f] = np.corrcoef(parameter_estimates[f,:,i], parameter_true[f,:,j])[1,0]
```

```python
corrmatrix = np.mean(corrmatrix,axis=2)
```

```python
# correlation between true and estimated parameter - recovery
fig, ax = plt.subplots(figsize=(15, 10))
annot = np.diag(np.diag(corrmatrix))
annot = np.round(annot,2)
annot = annot.astype('str')
annot[annot=='0.0']=''
sns.heatmap(corrmatrix, annot = annot, fmt='', ax=ax, annot_kws={"fontsize":24},cmap="YlGnBu",center=0)
ax.set_title("Parameter recovery",fontsize=18);
ax.set_xlabel("true parameters",fontsize=18);
ax.set_ylabel("re-estimated parameters",fontsize=18);
ax.set_xticklabels(parameter_labels, rotation=45);
ax.set_yticklabels(parameter_labels, rotation=0);
ax.tick_params(labelsize=18)
cbar = ax.collections[0].colorbar
# here set the labelsize by 20
cbar.ax.tick_params(labelsize=18)
cbar.ax.set_ylabel('Pearson correlation', size=20);
fig.tight_layout()
figpath = '/Users/jolandamalamud/phd/projects/gng_panda_antler/gng_panda/results/paper_figures/'
plt.savefig(figpath + 'parameter_recovery.svg', format='svg',transparent=True)
```
