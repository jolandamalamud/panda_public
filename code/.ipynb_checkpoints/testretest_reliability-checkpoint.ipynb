```python
filepath = '/Users/jolandamalamud/phd/projects/gng_panda_antler/gng_panda/data/'
panda = Panda(filepath,[])
# load raw task data
D = panda.load_data()
# prep action choices to plot
data = panda.extract_data(D)
```

```python
modelling = panda.load_modelling_results('modelling_results/', ['ll2b2a2epxb', 'llb'])
#transform_params = panda.transform_parameters(modelling['emmap'])
random_baseline_model = panda.load_modelling_results('modelling_results/', ['llb'])
# load modelling data\n",
modelling = panda.load_modelling_results('modelling_results/', ['ll2b2a2epxb', 'llb'])
#transform_params = panda.transform_parameters(modelling['emmap'])
random_baseline_model = panda.load_modelling_results('modelling_results/', ['llb'])
# load rct data
dfRCT = panda.load_rctdata()
```

```python
# define all variables of interest
parameter_labels = ['rew_se', 'loss_se', 'rew_LR', 'loss_LR', 'app_Pav', 'av_Pav', 'noise', 'bias', 'rbias']
parameter_labels_transformed = []
# for i in parameter_labels[:-2]: parameter_labels_transformed.append(i + '_trans')
gng_columns = parameter_labels + \
                ['exclusion', 'goprotot', 'gopro_g2w', 'gopro_g2a', 'gopro_ng2w',
                 'gopro_ng2a', 'acctot', 'acc_g2w', 'acc_g2a', 'acc_ng2w', 'acc_ng2a','iL']
gng_data = np.vstack((modelling['emmap'], random_baseline_model['emmap'], modelling['excluding'],
                       np.nanmean(data['a_go'],axis=(0,1)), np.nanmean(data['a_go'],axis=0),
                       np.nanmean(data['a_correct'],axis=(0,1)), np.nanmean(data['a_correct'],axis=0),
                       modelling['iL']))
rct_columns = ['rctid', 'group']
rct_data = dfRCT[['identifier_n', 'group']]
```

```python
df_panda = panda.create_panda_df(data, gng_columns, gng_data, rct_columns, rct_data)
df_panda = df_panda.astype(float)
```

```python
# no task data at all
noTask = df_panda[['acctot1','acctot2','acctot3']].isna().all(axis=1)
print('no task data: ' + str(sum(noTask)) + ' from that sertraline: '  + \
      str(df_panda['group'][noTask].sum()) + ', placebo: ' + str((df_panda['group'][noTask]==0).sum()))
print('-> N = ' + str(len(df_panda)-sum(noTask)) + ' from that sertraline: '  + \
      str(df_panda['group'][~(noTask)].sum()) + ', placebo: ' + str((df_panda['group'][~(noTask)]==0).sum()))

# drop participants with no task data
df_panda = df_panda[~noTask].reset_index(drop=True)
```

```python
trt_reliability = pd.DataFrame()
for j in gng_columns:
    x = [x for x in df_panda.columns if j in x]
    if j == 'bias': x = [z for z in x if 'r' not in z]
    y = df_panda[x].corr().to_numpy()
    trt_reliability[j] = [y[0,1],y[0,2], y[1,2]]
```

```python
# test retest realiability for all gng data and parameter estimates
panda_stats.get_trt_reliability(df_panda, gng_columns)
trt_reliability.rename(index={0: 'baseline to T = 2', 1: 'baseline to T = 6', 2: 'T = 2 to T = 6'})
```

```python
# only included task runs
condition = (df_panda['exclusion1']==0)&(df_panda['exclusion2']==0)&(df_panda['exclusion3']==0)
panda_stats.get_trt_reliability(df_panda, parameter_labels[:-1], condition)
```

```python
import rpy2.robjects as ro
import rpy2.robjects.numpy2ri
from rpy2.robjects.packages import importr
from rpy2.robjects import pandas2ri
rpy2.robjects.numpy2ri.activate()
psych = importr("psych")
tmp = []
for i in parameter_labels:
    B = df_panda[[i + '1',i + '2',i + '3']][condition].values
    nr,nc = B.shape
    Br = ro.r.matrix(B, nrow=nr, ncol=nc)
    icc = psych.ICC(Br)
    icc_df = pandas2ri.rpy2py(icc[0])
    tmp.append(icc_df.iloc[4])
pd.DataFrame(tmp, index=parameter_labels)
```

```python
df_icc = pd.DataFrame(tmp, index=parameter_labels)
df_icc = df_icc.drop('type', axis=1).round(2)
display(df_icc)
print(df_icc.to_latex())
```

```python
plot_time_label = ['baseline', 'week 2', 'week 6']
plot_parameter_label = ['reward sensitivity','loss sensitivity','reward learning rate','loss learning rate', \
                       'appetitive Pavlovian bias', 'aversive Pavlovian bias', 'noise', 'bias']

fig, ax = plt.subplots(2,4,figsize=(12, 6))
m,n = 0,0
for i,j in enumerate(parameter_labels[:-1]):
    x = [x for x in df_panda.columns if j in x]
    if j == 'bias': x = [z for z in x if 'r' not in z]
    rho = df_panda[x][condition].corr()
    pval = df_panda[x][condition].corr(method=lambda x, y: pearsonr(x, y)[1]) - np.eye(*rho.shape)
    p = pval.applymap(lambda x: ''.join(['*' for t in np.array([0.01,0.05,0.1]) /((len(parameter_labels[:-1])**2-1)/2) if x<=t]))
    annot = (rho.round(2).astype(str) + p).values
    for a in range(len(plot_time_label)): annot[a,a] = '1'
    matrix = np.triu(rho,k=1)
    g = sns.heatmap(rho,ax=ax[n,m], annot = annot, fmt='', cbar=False, mask=matrix, \
                cmap="YlGnBu", xticklabels=plot_time_label, vmin=-1, vmax=1)
    g.set_yticklabels(labels=plot_time_label, va='center')
    ax[n,m].set_title(plot_parameter_label[i], y=1, x=0.45, fontsize=12)
    m += 1
    if m > 3: m,n = 0,1

fig.tight_layout()
fig.suptitle("Pearson correlation between different time points", fontsize=16)
fig.subplots_adjust(top=0.85, hspace=0.5)

figpath = '/Users/jolandamalamud/phd/projects/gng_panda_antler/gng_panda/results/paper_figures/'
for i in ['eps', 'svg']:
    plt.savefig(figpath + 'test_retest_pearson.' + i, format=i,transparent=True)
    
```

```python
# parameter correlation
plot_parameter_label = ['reward\nsensitivity','loss\nsensitivity','reward\nlearning rate','loss\nlearning rate', \
                       'appetitive\nPavlovian bias', 'aversive\nPavlovian bias', 'noise', 'bias']
fig, ax = plt.subplots(1,3,figsize=(20, 6))
for j in range(1,4):
    tmp = []
    for i in parameter_labels[:-1]:
        tmp.append(i + str(j))
    rho = df_panda[tmp].corr()
    pval = df_panda[tmp].corr(method=lambda x, y: pearsonr(x, y)[1]) - np.eye(*rho.shape)
    p = pval.applymap(lambda x: ''.join(['*' for t in np.array([0.01,0.05,0.1]) /((len(parameter_labels[:-1])**2-1)/2) if x<=t]))
    matrix = np.triu(rho,k=1)
    annot = (rho.round(2).astype(str) + p).values
    for n in range(len(parameter_labels[:-1])): annot[n,n] = '1'
    sns.heatmap(rho,ax=ax[j-1], annot = annot, fmt='', cbar=False, mask=matrix,  vmin=-1, vmax=1, \
                cmap="YlGnBu", xticklabels=plot_parameter_label, yticklabels=plot_parameter_label)
    ax[j-1].set_title(plot_time_label[j-1],fontsize=12)
fig.suptitle("Pearson correlation between cognitive parameters", fontsize=16)
fig.subplots_adjust(top=0.8)
fig.tight_layout()
for i in ['eps', 'svg']:
    plt.savefig(figpath + 'parameter_correlation.' + i, format=i,transparent=True)
    
```
