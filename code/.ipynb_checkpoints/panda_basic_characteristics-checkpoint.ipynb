```python
#missing data related to baseline variables
baseline_continuous =  ['age', 'gad1log', 'phq1log', 'bdi1log']
baseline_categorical = ['site', 'cis', 'dep', 'education', \
                        'AD_past','sex', 'ethnic', 'fin', 'empstat', 'marstat', 'group']
tab_list = []
tab_list_reduced = []
for i in range(1,4):
    tab = []
    for j in baseline_continuous + baseline_categorical:
        model = smf.logit('missing_gngdata' + str(i) +' ~ ' + j, data=df_panda).fit(disp=False);
        tab.append([j, round(model.params[1],2), round(model.pvalues[1],3)])
    tab_list.append(pd.DataFrame(tab, columns=['baseline variable', 'estimate', 'pvalue']))
for i in tab_list:
    tab_list_reduced.append(i[i['pvalue']<0.05])
print('baseline variables related to missing data?')
disp.display_side_by_side(tab_list_reduced[0],tab_list_reduced[1],tab_list_reduced[2])
# print('multiple comparisons: ' + str(round(0.05/(3*len(baseline_continuous + baseline_categorical)),4)))
```

```python
# create dataframe for mixed-effects modelling
mle_df = pd.DataFrame(columns = gng_columns + rct_columns)
for c in gng_columns:
    mle_df[c] = pd.concat([df_panda[c + '1'], df_panda[c + '2'], df_panda[c + '3']], axis=0, ignore_index=True)
    
for c in rct_columns:
    mle_df[c] = pd.concat([df_panda[c], df_panda[c], df_panda[c]], axis=0, ignore_index=True)

mle_df['subject'] = np.tile(np.arange(len(df_panda)),3)
mle_df['weeks'] = np.hstack((np.tile(weeks[0], len(df_panda)),np.tile(weeks[1], len(df_panda)),np.tile(weeks[2], len(df_panda))))
mle_df['time'] = np.hstack((np.tile(0, len(df_panda)),np.tile(1, len(df_panda)),np.tile(2, len(df_panda))))
mle_df['group'] = np.hstack((np.tile(0, len(df_panda)),df_panda['group'],df_panda['group']))
mle_df['gadlog'] = np.hstack((df_panda['gad1log'],df_panda['gad2log'],df_panda['gad3log']))
mle_df['phqlog'] = np.hstack((df_panda['phq1log'],df_panda['phq2log'],df_panda['phq3log']))
```

```python
# does the group have an effect on go probability and accuracy?
import warnings
from statsmodels.tools.sm_exceptions import ConvergenceWarning
warnings.simplefilter('ignore', ConvergenceWarning)

timing = [((mle_df['time']==0)|(mle_df['time']==1)),((mle_df['time']==0)|(mle_df['time']==2)), \
          ((mle_df['time']==0)|(mle_df['time']==1)|(mle_df['time']==2))]
timing_label = ['T = 2','T = 6','over time']

measure = ['gopro', 'acctot']
for t in range(3):
    print('###############################################################################')
    print(timing_label[t] + ':')
    for j in measure:
        x = [x for x in gng_columns if j in x]
        for i in x:
            if i in ['acctot', 'acc_g2w', 'acc_g2a']: print(i, end='\t\t\t')
            else: print(i, end='\t\t')
            mle_df[i] = pd.to_numeric(mle_df[i])
            pstats.mle(i+' ~ group + site + cis + dep + time', mle_df[timing[t]],['group'], [])
```

```python
# does go probability and accuracy relate to symptoms?
psychiatric_questionnaire = ['gad', 'phq', 'bdi']
measure = ['gopro', 'acctot']
for p in psychiatric_questionnaire:
    print('###############################################################################')
    print(p + ':')
    for t in range(3):
        print('###############################################################################')
        print(timing_label[t] + ':')
        for j in measure:
            x = [x for x in gng_columns if j in x]
            for i in x:
                pstats.mle(p + 'log ~ ' + i + ' + group + time', mle_df[timing[t]],[i], i)
```
