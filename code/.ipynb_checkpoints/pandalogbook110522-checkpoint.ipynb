```python
import scipy.io
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from plotting import Plotting
from loading_preparing_data import Panda
```

```python
# load raw data
filepath = '/Users/jolandamalamud/phd/projects/gng_panda_antler/gng_panda/data/'
panda = Panda(filepath)
D = panda.load_data()
```

New model with separate learning rate for trials where subjects experience the association between go action and reward (lr + kappa) and where they experience the association between nogo action and punishment (lr - kappa) fits the data much better as it can account for the mean low go probability in the g2w (Ouden et al. 2017, eLife). Here only fitted to data from session 2 (2 weeks).

```python
# prep action choices to plot
data = panda.extract_data(D)
# load surrogate data
surrogate_data = panda.load_surrogate_data('modelling_results/sess2/', 'llb2aepxbgng')
data_incl = panda.extract_data(D[data['sess']==2])
data_incl = panda.extract_surrogate_data(data_incl, surrogate_data)
opt_plot = {"xlabel": 'trial',
            "ylabel": 'probability to go',
            "title": ['go to win', 'go to avoid', 'nogo to win', 'nogo to avoid'],
            "surr_to_plot": data_incl['surr'],
            "format": False}
data_to_plot = data_incl['a_go']
p = Plotting()
p.error_shaded_line_plot(data_to_plot, opt_plot)
p.goprobability_scatter_plot(data_incl, opt_plot)
```

```python
modelling = panda.load_modelling_results('/modelling_results/sess2/', ['llb2aepxbgng', 'llb'])
print('learning difference parameter averaged over participants (SD): ' + str(round(np.nanmean(modelling['emmap'][-1,:]),2)) + ' (' + str(round(np.nanstd(modelling['emmap'][-1,:]),2)) + ')')
print('only included participants (SD): ' + str(round(np.nanmean(modelling['emmap'][-1,~modelling['excluding']]),2)) + ' (' + str(round(np.nanstd(modelling['emmap'][-1,~modelling['excluding']]),2)) + ')')
print('high variability but on average negative meaning the majority learns better from punishment and nogo')
```

```python
# exclude participants where the random bias model explains the data better
opt_plot["excluding"] = modelling["excluding"]
opt_plot["legend"] = ['empirical', 'model']
p.error_shaded_line_plot(data_to_plot, opt_plot)
p.goprobability_scatter_plot(data_incl, opt_plot)
```

```python
# trial data
dfRCT = panda.load_rctdata()
rctids = dfRCT['identifier_n']
group = dfRCT['group']
taskids = np.unique(data_incl['subids'])
```

```python
# compare parameter in group
group_matched = panda.match_ids(group, taskids, rctids)
sert = modelling["emmap"][:,group_matched == 2]
placebo = modelling["emmap"][:,group_matched == 1]
sert = modelling["emmap"][:,np.logical_and(~modelling["excluding"], group_matched == 2)]
placebo = modelling["emmap"][:,np.logical_and(~modelling["excluding"], group_matched == 1)]
print('mean and SD of cognitive parameters and pvalue of difference')
parameter_labels = ['outcome sensitivity', 'learning rate', 'pavlovian bias', 'noise', 'bias', 'lr difference']
panda.ttest2(sert, placebo, parameter_labels)
```

No significant difference between groups in any cognitive parameter...

```python
opt_plot["excluding"] = modelling["excluding"]
opt_plot['group'] = group_matched
opt_plot["legend"] = ['placebo empirical', 'placebo model','sertraline empirical','sertraline model']
p.group_line_plot(data_to_plot, opt_plot)
```

```python
# relation to psychiatric score
weeks = ['2', '6', '12']
for i in weeks:
    gad = dfRCT['log_gadtot_'+i+'wk']
    gad_matched = panda.match_ids(gad, taskids, rctids)
    print('association between...')
    panda.correlation(modelling['emmap'][:,~modelling['excluding']], \
                      gad_matched[~modelling['excluding']], parameter_labels, 'total gad at '+i+ ' weeks')
```
