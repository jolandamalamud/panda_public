```python
import pyreadstat
```

```python
# Load RCT data file
dtafile = '../../data/trial_data/data_merged_clean_13.dta'
df, meta = pyreadstat.read_dta(dtafile)
```

```python
df.info()
```

```python
# Print all columns
for i in df.columns:
    print(i)
```

```python
# Specify variables of interest in RCT
df_save = df[['identifier_n', 'group'] + [i for i in ['gadtot', 'phqtot', 'becktot', 'phq1']] + \
                 [i + '_' + j + 'wk' for i in ['log_gadtot', 'log_phqtot', 'log_becktot', 'phq1'] \
                  for j in ['2', '6', '12']] + \
                 ['_site_n', 'cistotal_cat', 'depr_dur_2years', 'age', 'edu3', 'antidepressantsinpast', \
                  'sex', 'ethnic', 'fin3', 'empstat2', 'marstat3', 'cisdepscore']]

# Rename column names
df_save.columns = ['ID', 'group'] + [i + '0' for i in ['gad', 'phq', 'bdi', 'anh']] + \
                [i + str(j) + 'log' for i in ['gad', 'phq', 'bdi'] for j in range(1,4)] + \
                ['anh' + str(j) for j in range(1,4)] + ['site', 'cis', 'dep', 'age', 'education', 'AD_past', \
               'sex', 'ethnic', 'fin', 'empstat', 'marstat', 'cisscore']
```

```python
# Save DataFrame
df_save.to_csv('../../data/trial_data/gng_panda_data.csv')
```

Description of variables in RCT:
- CIS-R total score (1=0-11, 2=12-19; 3=>=2-49)
- CIS-R depression duration (years) (1=<2years;2=>=2years)
- site (1=Bristol;2=Liverpool;3=York;4=London)
- marital status (1=married;2=single;3=separated/divorced/widowed)
- age (years)
- depression in the past (1=no;2=yes)
- highest educational qualification (1=A Level or higher;2=GCSE,standard grade or other;3=no formal qualification)
- antidepressants in the past (1=no;2=yes)
- financial difficulty (1=living comfortably or doing alright; 2=just about getting by; 3=finding it difficult or very difficult)
- ethinicity (1=white;2:7=ethnic minority)
- employment status (1=in paid employment;2=not employed)
- sex (1=male;2=female)
