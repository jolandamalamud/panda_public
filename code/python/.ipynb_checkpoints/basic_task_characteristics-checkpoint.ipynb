```python
import numpy as np
import pandas as pd

from loading_preparing_data import PrepData
from gng_statistics import GNGstats
from plotting import Plotting

import seaborn as sns; import matplotlib.pyplot as plt
sns.set(style='white',font_scale=1.5)

import statsmodels.formula.api as smf
import warnings 
from statsmodels.tools.sm_exceptions import ConvergenceWarning
warnings.simplefilter('ignore', ConvergenceWarning)
import scipy
from scipy.stats import ttest_ind
```

```python
# Set the file path
filepath = '../../'

# Create an instance of the PrepData class
prepdata = PrepData('panda', filepath, ['gad', 'phq', 'bdi'])
# Create an instance of the GNGstats class
gngstats = GNGstats()
# Create an instance of the Plotting class
plotting = Plotting('panda')

# Load raw task data
D = prepdata.load_data()
# Prepare action choices for plotting
data = prepdata.extract_data(D)
# Load model fits
modelling = prepdata.load_modelfits()

# Load rct data
dfRCT = prepdata.load_rctdata()

# Set the file path for saving figures related to the paper
figpath = filepath + 'results/paper_figures/'
```

```python
# Define all variables of interest from RCT
baseline_vars = ['site', 'cis', 'dep', 'age', 'education', 'AD_past', \
               'sex', 'ethnic', 'fin', 'empstat', 'marstat', 'cisscore']

# Questionnaire variables which need to be log-transformed
log_transfer = [i + '0' for i in prepdata.psychiatric_questionnaire] + [i for i in dfRCT.columns if 'anh' in i]

# Create DataFrame
df_panda = prepdata.create_df(data, modelling, dfRCT, log_transfer)
df_panda = df_panda.astype(float)
```

---------------------------------------------------------------------------------------------------------------
Missing Data
--

```python
# Information for trial consort 
trial_consort = pd.DataFrame()
for i in range(4):
    for g in range(2):
        trial_consort.loc[i,g] = sum(~df_panda['gad'+str(i)+'log'][df_panda['group']==g].isna())
        if i !=3: trial_consort.loc[i,g+2] = sum(~df_panda['acctot' + str(i)][df_panda['group']==g].isna())
        if i !=3: trial_consort.loc[i,g+4] = sum(df_panda['exclusion' + str(i)][df_panda['group']==g]==0)
trial_consort.columns = [3 * ['placebo'] + 3 * ['sertraline'], 2*['gad', 'gng', 'informative']]
trial_consort.index = plotting.time_label + ['week 12']
trial_consort
```

```python
# Check if there is any task data available
noTask = df_panda[['acctot0', 'acctot1', 'acctot2']].isna().all(axis=1)

# Print the count of participants with no task data and the breakdown by treatment group
print('Number of participants with no task data: ' + str(sum(noTask)) +
      ', Sertraline: ' + str(df_panda['group'][noTask].sum()) +
      ', Placebo: ' + str((df_panda['group'][noTask] == 0).sum()))

# Print the count of participants with task data and the breakdown by treatment group
print('N = ' + str(len(df_panda) - sum(noTask)) +
      ', Sertraline: ' + str(df_panda['group'][~noTask].sum()) +
      ', Placebo: ' + str((df_panda['group'][~noTask] == 0).sum()))

# Drop participants with no task data from the DataFrame
df_panda = df_panda[~noTask].reset_index(drop=True)
```

```python
# Missing data due to missing task or gad-7 data (c.f. Trial Consort)
for t in range(3):
    df_panda['missing_gngdata' + str(t)] = df_panda['acctot' + str(t)].isna().astype(int)
    print('missing task data at sessions ' + str(t) + ': ' + str(sum(df_panda['missing_gngdata' + str(t)])))
    df_panda['missing_gad' + str(t)] = df_panda['gad' + str(t) + 'log'].isna().astype(int)
    print('missing GAD-7 data at sessions ' + str(t) + ': ' + str(sum(df_panda['missing_gad' + str(t)])))
    df_panda['missing' + str(t)] = df_panda['acctot' + str(t)].isna() | df_panda['gad' + str(t) + 'log'].isna()
    print('missing overall ' + str(t) + ': ' + str(sum(df_panda['missing' + str(t)])))
```

```python
# Missing task data related to baseline variables
baseline_continuous =  ['age', 'gad0log', 'phq0log', 'bdi0log']
baseline_categorical = [i for i in baseline_vars if i != 'age'] + ['group']
tab_list = []
tab_list_reduced = []
for i in range(3):
    tab = []
    for j in baseline_continuous + baseline_categorical:
        model = smf.logit('missing_gngdata' + str(i) +' ~ ' + j, data=df_panda).fit(disp=False);
        tab.append([j, round(model.params[1],2), round(model.pvalues[1],3)])
    tab_list.append(pd.DataFrame(tab, columns=['baseline variable', 'estimate', 'pvalue']))
for i in tab_list:
    tab_list_reduced.append(i[i['pvalue']<0.05])
print('baseline variables related to missing data?')
plotting.display_side_by_side(tab_list_reduced[0],tab_list_reduced[1],tab_list_reduced[2])
```

```python
# To determine the site with a higher amount of missing task data, a post-hoc chi-square (X2) test was conducted
sitetest = gngstats.chi2_test(df_panda['site'], df_panda['missing_gngdata1'], ['missing', 'not missing'])
sitetest.index = ['Bristol','Liverpool','York','London']
sitetest
```

655 patients were recruited and randomly assigned to sertraline (326, 50\%) and
placebo (329, 50\%). Two patients in the sertraline group did not complete a substantial proportion of the baseline assessment and were excluded. Additionally, 25 patients (9 from the sertraline group and 16 from placebo) did not complete the Go/NoGo task at any time-point. This left 628 participants (315 sertraline and 313 placebo) for analyses. Task data for 7 patients at baseline, 99 patients at 2 weeks, and 145 patients at 6 weeks were missing. 
Missing follow-up data were more common in
participants who had higher baseline depressive and anxiety symptoms, financial difficulties, 
were from ethnic minorities and recruited from London. 
Missing data did not differ statistically by treatment allocation.

```python
# Accuracy differs between Pavlovian congurent (go2win & nogo2avoid) and incongurent(nogo2win & go2avoid) condition 
# for each session using independent T-test
# 1) accuracy sig different for same action different valence
# 2) accuracy sig different for same valence different action
for t in range(3):
    print('T = ' + plotting.time_label[t] + ':')
    print('go2avoid vs go2win: ' + str(ttest_ind(df_panda['acc_g2a' + str(t)].dropna(), \
                    df_panda['acc_g2w' + str(t)].dropna(), equal_var=False)))
    print('nogo2avoid vs nogo2win: ' + str(ttest_ind(df_panda['acc_ng2a' + str(t)].dropna(), \
                    df_panda['acc_ng2w' + str(t)].dropna(), equal_var=False)))
    print('go2avoid vs nogo2avoid: ' + str(ttest_ind(df_panda['acc_g2a' + str(t)].dropna(), \
                    df_panda['acc_ng2a' + str(t)].dropna(), equal_var=False)))
    print('go2win vs nogo2win: ' + str(ttest_ind(df_panda['acc_g2w' + str(t)].dropna(), \
                    df_panda['acc_ng2w' + str(t)].dropna(), equal_var=False)))
```

Participants were better at learning to press the button in anticipation of reward, and better at withholding a response in anticipation of punishment. 

```python
# Differential accuracy between Pav con and incon condition differs between actions
print('"Pav" different between go and nogo:')
for t in range(3):
    print('T = ' + plotting.time_label[t] + ':')
    print(ttest_ind((df_panda['acc_g2w' + str(t)]-df_panda['acc_g2a' + str(t)]).dropna(), \
                    (df_panda['acc_ng2a' + str(t)]-df_panda['acc_ng2w' + str(t)]).dropna(), equal_var=False))
print(100*'-')

# Differential accuracy between Pav con and incon condition differs between valence
print('"Pav" different between win and loss:')
for t in range(3):
    print('T = ' + plotting.time_label[t] + ':')
    print(ttest_ind((df_panda['acc_g2w' + str(t)]-df_panda['acc_ng2w' + str(t)]).dropna(), \
                    (df_panda['acc_ng2a' + str(t)]-df_panda['acc_g2a' + str(t)]).dropna(), equal_var=False))
```

- Larger difference between Pavlovian congurent and incongurent condition where nogo action is required
- Larger difference between Pavlovian congurent and incongurent in loss than win condition at baseline


```python
# Differential accuracy between Pav con and incon condition for each session does NOT differ between groups
for i in range(1,3):
    g2w_g2a = df_panda['acc_g2w' + str(i)]-df_panda['acc_ng2w' + str(i)]
    ng2a_ng2w = df_panda['acc_ng2a' + str(i)]-df_panda['acc_g2a' + str(i)]
    g2w_ng2w = df_panda['acc_g2w' + str(i)]-df_panda['acc_ng2w' + str(i)]
    ng2a_g2a = df_panda['acc_ng2a' + str(i)]-df_panda['acc_g2a' + str(i)]
    print('T = ' + plotting.time_label[i] + ':\ng2w - g2a: ' + \
          str(ttest_ind(g2w_g2a[df_panda['group']==1].dropna(), \
                    g2w_g2a[df_panda['group']==0].dropna(), equal_var=False))  + \
          '\nng2a - ng2w: ' + str(ttest_ind(ng2a_ng2w[df_panda['group']==1].dropna(), \
                    ng2a_ng2w[df_panda['group']==0].dropna(), equal_var=False)) + 
          '\ng2w - ng2w: ' + str(ttest_ind(g2w_ng2w[df_panda['group']==1].dropna(), \
                    ng2a_ng2w[df_panda['group']==0].dropna(), equal_var=False)) + 
          '\nng2a - g2a: ' + str(ttest_ind(ng2a_g2a[df_panda['group']==1].dropna(), \
                    ng2a_ng2w[df_panda['group']==0].dropna(), equal_var=False)))
```

Mixed-Effects Modelling
--

```python
mle_df = prepdata.create_mle_df(df_panda, prepdata.gng_variables, baseline_vars)
timing = [mle_df['time']<2,((mle_df['time']==0)|(mle_df['time']==2)), mle_df['time']<3]
timing_label = ['T = 2','T = 6','over time']
```

```python
# Does the group have an effect on go probability and accuracy?
measure = ['gopro', 'acctot']
for t in range(3):
    print('###############################################################################')
    print(timing_label[t] + ':')
    for j in measure:
        x = [x for x in prepdata.gng_variables if j in x]
        for i in x:
            if i in ['acctot', 'acc_g2w', 'acc_g2a']: print(i, end='\t\t\t')
            else: print(i, end='\t\t')
            mle_df[i] = pd.to_numeric(mle_df[i])
            gngstats.mle(i+' ~ group + site + cis + dep + time', mle_df[timing[t]],['group'], [])
```

```python
# Does go probability and accuracy relate to symptoms?'
measure = ['gopro', 'acctot']
for p in prepdata.psychiatric_questionnaire:
    print('###############################################################################')
    print(p + ':')
    for t in range(3):
        print('###############################################################################')
        print(timing_label[t] + ':')
        for j in measure:
            x = [x for x in prepdata.gng_variables if j in x]
            for i in x:
                gngstats.mle(p + 'log ~ ' + i + ' + group + time', mle_df[timing[t]], \
                             [i], i)
```

```python
# Baseline variables related to GAD-7 (control for those variables when looking at symptom links)
rho = mle_df[['gadlog', 'phqlog', 'bdilog', 'age'] + baseline_categorical].corr()
pval = mle_df[['gadlog', 'phqlog', 'bdilog', 'age'] + baseline_categorical].corr(method=lambda x, y: \
                                                    scipy.stats.pearsonr(x, y)[1]) - np.eye(*rho.shape)
p = pval.applymap(lambda x: ''.join(['*' for t in [0.01,0.05,0.1] if x<=t]))
rho.round(2).astype(str) + p
```

age, financial difficulties and startification variables are related to BDI
